```python
def clear_text(text, lowerize=True):
    # чистим все, что не входит в задданый компайл
    pat = re.compile(r'[^A-Za-z0-9 \-\n\r.,;!?А-Яа-я]+')
    cleared_text = re.sub(pat, ' ', text)

    # переводим в нижний регистр
    if lowerize:
        cleared_text = cleared_text.lower()
    # токенизируем текст
    tokens = cleared_text.split()
    return tokens

def make_substrings(tokens, k=4):
    # передаем подстроку из k токенов
    for i in range(max(len(tokens) - k + 1, 1)):
        yield ' '.join(tokens[i:i + k])

def text_to_topic(texts, topics):
    matrix = np.empty((0, len(topics)), float)
    texts_tokens = [clear_text(t) for t in texts]
    topics_syn = []
    for topic_key in topics:
        topic_syn = [' '.join(topic_key.split()[1:])]
        if topics[topic_key] is not None:
            topic_syn += topics[topic_key]
        topics_syn.append(topic_syn)
        
    for text_tokens in texts_tokens:
        ast = base.AST.get_ast(list(make_substrings(text_tokens)))
        scores = []
        for topic_syn in topics_syn:
            score = 0
            for syn in topic_syn:
                score = max(ast.score(syn), score)
            scores.append(score)
        row = np.array(scores)    
        matrix = np.append(matrix, [row], axis=0)
    return matrix
```

```python
def clear_keyword(keyword, lowerize=True):
    # чистим все, что не входит в задданый компайл
    pat = re.compile(r'[^A-Za-z0-9 \-\n\r.,;\'!?А-Яа-я]+')
    cleared_keyword = re.sub(pat, ' ', keyword)

    # переводим в нижний регистр
    if lowerize:
        cleared_keyword = cleared_keyword.lower()
    return cleared_keyword.split(', ')

def keyw_to_topic(keywords, topics):
    matrix = np.empty((0, len(topics)), float)
    keywords_tokens = []
    for t in keywords:
        if t is not np.nan:
            keywords_tokens.append(clear_keyword(t))
        else:
            keywords_tokens.append([''])
    topics_syn = []
    for topic_key in topics:
        topic_syn = [' '.join(topic_key.split()[1:])]
        if topics[topic_key] is not None:
            topic_syn += topics[topic_key]
        topics_syn.append(topic_syn)

    for keyword_tokens in keywords_tokens:
        scores = []
        for topic_syn in topics_syn:
            score = 0
            for syn in topic_syn:
                for keyword in keyword_tokens:
                    a = set(syn.split())
                    b = set(keyword.split())
                    c = len(a.intersection(b))/len(a.union(b))
                    score = max(score, c)
            scores.append(np.sqrt(score))
        row = np.array(scores)    
        matrix = np.append(matrix, [row], axis=0)
    return matrix
```

```python
def topic_to_topic(R, threshold):
    if threshold is not None:
        ost = []
        n_v = np.sum(R > threshold, axis=1)
        for i in range(len(n_v)):
            if n_v[i] == 0:
                ost.append(i)
        R = np.delete(R, ost, axis=0)
        n_v = np.delete(n_v, ost)
        w = 1 / n_v
        A = R.T.dot(np.diag(w)).dot(R)
        return A, ost
    else:
        A = R.T.dot(R)
    return A
```
